## Plain vanilla option pricing

All'interno del notebook viene simulata l'evoluzione del prezzo di asset, $s(t)$ nel tempo, verificando la validità della teoria fornita da **Black-Scholes** per l' *Option Pricing*. La teoria prevede il corretto prezzamento delle opzioni $europee$ di tipo Call e Put.

La teoria ipotizza che l'evoluzione del prezzo di un asset nel tempo, $S(t)$, sia stocastica e possa essere descritta mediante un *Moto Browniano Geometrico (GBM)* con volatilità $\sigma$ e drift $\mu = r$, dove $r$ è il tasso di interesse costante.

Un'opzione **europea** fornisce al suo *compratore* il diritto (ma non l'obbligo) di comprare (opzione $call$) o vendere (opzione $put$) ad un prezzo prefissato $K$ (detto prezzo d'esercizio o prezzo strike)  una particolare attività finanziaria *ma* con facoltà di vendere o comprare (se il prezzo d'esercizio sarà vantaggioso rispetto al prezzo di mercato corrente dell'attività in questione e al premio) limitata alla sola data di scadenza $T$ (delivery time).

Il **profitto** sarà possibile solo se, alla scadenza di $T$, il prezzo di mercato $S(T)$ dell'asset è maggiore di $K$. Si possono generare quindi sue situazioni: 
1. $S(T) > K$ : Il compratore esercita l'opzione di $Call$ (comprare) , con profitto $S-K$
2. $S(T) < K$ : Il compratore esercita l'opzione di $Put$ (vendere), con "profitto" $K-S$


Nella teoria di *Black-Scholes* , l'asset ha un prezzo $S\sim GBM(\mu,\sigma^2)$, con valore iniziale $S(0)$, e evolve nel tempo come: 

$$S(t) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\sqrt{t}\right]$$   
dove $W(t)\sim\mathcal{N}(0,t)$. Questa casistica la chiameremo `Campionamento Diretto `.
    
    
Inoltre, siccome gli incrementi di  $W$ sono indipendenti e distribuiti in modo normale, il GBW può essere discretizzato senza commettere errori in  $t_1 < t_2 < … < t_n$ come :

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$

con $Z_i \sim \mathcal{N}(0,1)$ e $t_i$ istanti successivi nell' intervallo $[0,t]$. Questa casistica la chiameremo `Campionamento Discreto `.


I prezzi da assegnare alla *Call* e *Put option* , al tempo $T$, sono i seguenti:
- $C[S(T),T] = e^{-rT} max[0,S(T) - K]$
- $P[S(T),T] = e^{-rT} max[0,K-S(T)]$

dove $K$ è lo strike price e il termine $e^{-rT}$ tiene conto del tasso d'intresse *risk-free* presente.

All' interno del notebook saranno utilizzati i seguenti *parametri* : 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

L'obbiettivo è di calcolare il prezzo della *opzione di Call* e della *opzione di Put* al tempo $t = 0$ , dopo aver simulato il prezzo finale $S(T)$. 

Il calcolo è fattomediante due metodologie differenti:
1. Campionamento **Diretto** del prezzo finale dell'asset $S(T)$ per un $GBM(r,\sigma^2)$.
2. Campionamento del prezzo finale dell'asset $S(T)$, per tempi discreti, mediante **ricorsione**

### Exercise 03.1
Programma nella cartella `Ex-3`.

Il prezzo finale dell'asset $S(T)$ è stato campionato direttamente, in analogia a un moto $S\sim GBM(\mu,\sigma^2)$( per formula esatta, vedere paragrafo introduttivo).

L' analisi statistica è stata svolta mediante l'utilizzoo del *Data Blocking*; sono stati svolti $ M = 10^4$ campionamenti , suddivisi in $N = 100$ blocchi.

**Osservazione**: nel calcolo della media all' interno del Data Blocking, devo considerare anche i casi in cui $C(T) = 0$ e $P(T) = 0$, altrimenti l'analisi statistica risulterebbe incompleta in quanto considererebbe solo i casi favorevoli alla *Call/Put*.

I risultati ottenuti soo contenuti nei file :`Risultati_C.dat` e `Risultati_P.dat`.

```python
# ======= direct: CALL and PUT ========


fig=plt.figure(figsize=(17,6))
plt.suptitle("European Option Pricing - Campionamento Diretto", fontsize = 25)

plt.subplot(1,2,1)
blocks_C ,aveC, errC = np.loadtxt("Ex-3/Risultati_C.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(blocks_C, aveC, yerr = errC)
plt.ylabel('Call option price', fontsize = 15)
plt.xlabel('Block number', fontsize = 15)
plt.grid(True)
plt.axhline(y=14.975790778311286, color = 'r', linestyle = '-', label = "Risultato analitico\n14.976")
plt.legend()

plt.subplot(1,2,2)
blocks_P,aveP, errP = np.loadtxt("Ex-3/Risultati_P.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(blocks_P, aveP, yerr = errP)
plt.ylabel('Put option price', fontsize = 15)
plt.xlabel('Block number', fontsize = 15)
plt.grid(True)
plt.axhline(y=5.4595325819072364, color = 'r', linestyle = '-', label = "Risultato analitico\n5.4595")
plt.legend()

plt.show()

print('Risultato finale Call: ')
print(aveC[-1])
print('Incertezza statistica : ')
print(errC[-1])

print('\nRisultato finale Put: ')
print(aveP[-1])
print('Incertezza statistica : ')
print(errP[-1])
```

I risultati ottenuti sono compatibili con  quelli Analitici entro ≈ $1\sigma$. Anche aumentando il numero di campionamenti totali, la convergenza non migliora.
