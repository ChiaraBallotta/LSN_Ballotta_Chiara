```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

Durante il laboratorio è stato risolto il problema del **commesso viaggiatore**(TSP) mediante un **algoritmo genetico**.

Il *TSP* consinste in un commesso che deve visitare una e una sola volta $N$ città con posizione {$\vec{c_i}$}.

L'obbiettivo del problema è minimizzare la lunghezza totale del percorso, tenendo in considerazione che dovrà ritornare sempre alla città da cui è partito. Il problema dovrà quindi minimizzare una *funzione di loss*:
$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$

(dove $\vec{x}_{N+1}=\vec{x}_1$) è una particolare permutazione, P, del'ordine con cui le città  $\vec{c}_1, \vec{c}_2, \dots, \vec{c}_N$ sono visitate, i.e. 

$$
\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N = \vec{c}_{P1}, \vec{c}_{P2}, \dots, \vec{c}_{PN}
$$

Nel **codice**, l' ordine delle città visitate è presentato come un  `array `, al cui interno sono presenti i *label* che identificano ogni città. L' ordine con cui si presentano i *label* indica l'ordine con cui le città sono visitate.

Per ridurre la degenerazione della soluzione, si è deciso di fissare il primo elemento dell' array sempre a $1$.


Il problema è stato risolto mediante l'utilizzo di un algoritmo genetico. Di seguito è spiegato il funzionamento.

## Algoritmo Genetico - spiegazione

A grandi linee un **GA** funziona nel seguente modo:
1. Considera un insieme di percorsi, che prendono il nome di **individui**. L' insieme dei percorsi prende il nome di **Popolazione**

2. Valuta  la bontà di ciascun individuo in funzione della metrica selezionata

3. Mediante un operatore di **selezione**, estrae degli individui e opera delle **Mutazioni** che daranno vita alla una nuova **Generazione**. Le mutazioni possono essere di due tipologie:
    - Crossover
    - Permutazioni
    

4. Ripete i processi *1-3* per un numero di volte selezionato a priori.

## Algoritmo Genetico - implementazione

L' implementazione dell' algoritmo genetico avviene all' interno del file `Ex-9/popolazione.cpp`. All'interno del file sono definite $2$ classi:
- `Individuo`: ogni individuo contiene l' ordine con cui le cità sono visitate.

- `Popolazione` : contiene l' insieme di individui e implementa le funzioni di Mutazione.

### Individuo 
Popolazione Contiene l'ordine in cui le città sono visitate, salvate su un array `m_ind`. 

Un individuo può essere generato in maniera **Random** , pur rispettando le caratteristiche fissate dal problema, oppure a partire da un **vector**.

I **metodi** presenti nella classe sono i seguenti:
- `Individuo_new` : modifica l' individuo stesso generandone uno nuovo, in modo random.

- `Set_lenght` : Calcola la funzione di *Loss* e la salva nella variabile interna `m_lenght`.

- `Modif_city` : sostituisce la città selezionata, con una nuova

Ho creato una funzione di **check** `check` che controlla che  l' individuo in input rispetti le caratteristiche fissate dal problema, ovvero:
- La prima città visitata deve sempre essere $1$.
- Ogni città può essere visitata una e una sola volta.

La funzione inoltre riporta esplicitamente l' errore che riscontra e che invalida l' individuo ( questa caratteristica si è rivelata molto utile durante l' implementazione delle mutazioni, per capire nel dettaglio gli errori presenti)

### Popolazione

La classe contiene $N_{individui}$ salvati in un vector `m_popolaz`.

Una *popolazione* può essere generata in modo **Random**: ogni individuo sarà generato in modo *Random*; oppure può essere generata a seguito di mutazioni.
Ogni volta che viene generata una nuova *popolazione*, gli individui sono organizzati in ordine **decrescente** rispetto alla funzione di *Loss*. Il compito è svolto dalla funzione di `sorting`.

I **metodi** presenti nella classe sono i seguenti:
- `Set_individuo`: sostituisce un individuo prescelto con quello in *input*.
$$$$
- `selez`: Seleziona gli individui che andranno a subire le mutazioni. Per come è stata costruita la funzione di *Mutazione*, è nell' interesse del buon funzionamento dell' GA che vengano selezionati solamente i **migliori** individui per subire una mutazione; trovandoci comunque in un framework statistico, l'*individuo* $j-esimo$ che viene selezionato è estratto secondo la seguente legge di potenza: $$j= int(M \times r^p)+1$$ dove $r$ è un numero appartenente a una distribuzione random uniforme $r\in[0,1)$ e $p = 6$ ( valore da me scelto dopo aver osservato come variano i valori estratti in funzione di $p$). Questo metodo "predilige" i valori più bassi, e dunque gli *individui* con *Loss* più piccola.
$$$$
**Permutazioni**: sono state implementate diversi tipi di permutazione. Durante la fase di *Mutazione* , ogni *Permutazione* ha una probabilità $p^{(i)}_m$ (tipicamente $p^{(i)}_m < 10\%$, tranne che per il crossover dove è attorno al $50\%$) di essere chiamata. Il valore della probabilità $p^{(i)}_m$ è fissato nel file `input.h`. Le permutazioni implementate sono le seguenti:

    1. `Pair_permutaz`: permuta due città all' iterno dell' Individuo
    
    2. `Permutaz_N` : permuta un blocco di $N$ città
    
    3. `Shift` : Shifta m città, di n posizioni
    
    4. `Inversion` : inverte l'ordine di m città
    
    5. `Crossover` : esegui il crossover tra due individui
$$$$
- **Mutazione** esegue le permutazioni della popolazione. La *mutazione* avviene secondo il seguente schema:
    1. Selezione di $2$ individui, con il metodo descritto sopra.
    2. Gli individui selezionati sono salvati su una **nuova popolazione**
    3. Le operaziondi di *permutazione* sono eseguite sugli individui selezionati. Ogni qual volta avviene una permutazione, i figli sono salvati sulla nuova popolazione.
    4. Ripeto i passaggi *1-3* finche la * nuova popolazione* è stata riempita con il numero necessario di Individui.
    5. Esecuzione del `sorting`.

Il processo di **Mutazione** ha seguito varie mutazioni in corso d'opera, ma alla fine il metodo scelto funziona a dovere. Inizialmente avevo scelto di utilizzare una tecnica *Steady-state GA* dove la mutazione agiva man mano anche sui nuovi individui; per semplicità di calcolo l' ho sostituita con una tecnica di *Solution Generational GA* in cui i nuovi individui non sono utilizzati finchè l' intera popolazione non è stata rimpiazzata.

Di seguito sono illustrati il **migliore** percorso trovato dall'algoritmo assieme al **secondo** migliore. 

```python
N_generazioni = 500
N_cities = 34
best_best =np.loadtxt("Ex-9/Circolare/Best_best.dat",delimiter = None, unpack = 'true') 
best_second = np.loadtxt("Ex-9/Circolare/Best_second.dat",delimiter = None, unpack = 'true') 
n,x,y= np.loadtxt("Ex-9/Circolare/cities.dat", usecols = (0,1,2), delimiter = None, unpack = 'true')

x_best = np.zeros(N_cities)
y_best= np.zeros(N_cities)
contatore = 0
x_second = np.zeros(N_cities)
y_second = np.zeros(N_cities)
contatore_s = 0

lenght_best = best_best[0]
lenght_second = best_second[0]

for i in range(0,34,1):
    contatore = int(best_best[i+1])
    contatore_s = int(best_second[i+1])
    x_best[i] = x[contatore-1]
    y_best[i] = y[contatore-1]
    x_second[i] = x[contatore_s-1]
    y_second[i] = y[contatore_s-1]
    
fig = plt.figure() 

# Figure size

fig.set_size_inches(12,10)

#punti 2 e 15, selezionati manualmente
plt.plot(0.490951,  0.871187, marker ="*", color = "black", markersize = 20, label = "15", linestyle = "")
plt.plot(0.482205,  0.876058, marker ="*", color = "m", markersize = 20, label = "2", linestyle = "")

plt.plot(a,b,color = 'lightgrey')
plt.plot(x_best,y_best,marker = 'o',color = 'darkblue',label='Best path, lenght = '+ str(lenght_best ),linewidth=4)
plt.plot(x_second,y_second,marker = '.',color = 'orange',label='Second Best path, lenght = '+ str(lenght_second))   
plt.xlabel('x')
plt.ylabel('y')
plt.title("Disposizione Circolare ")
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True) 
plt.legend()



plt.show()
```

Il **second best path** e il **best path** differiscono solo per l'inversione di due punti molto vicini tra di loro( evidenziati con due stelline colorate). 

Siccome la differenza tra i due *path* è minima,è stato aggiunto manualmente  anche un **third best path** per avere una visuale migliore.

```python
best_third = np.loadtxt("Ex-9/Circolare/Best_third.dat",delimiter = None, unpack = 'true') 
lenght_third = best_third[0]
x_third = np.zeros(N_cities)
y_third = np.zeros(N_cities)
contatore_t = 0
for i in range(0,34,1):
    contatore_t = int(best_third[i+1])
    x_third[i] = x[contatore_t-1]
    y_third[i] = y[contatore_t-1] 

fig = plt.figure() 

# Figure size

fig.set_size_inches(12,10)

#punti 2 e 15, selezionati manualmente
plt.plot(a,b,color = 'lightgrey')
plt.plot(x_best,y_best,marker = 'o',color = 'blue',label='Best path, lenght = '+ str(lenght_best ),linewidth=4)
plt.plot(x_third,y_third,marker = '.',color = 'orange',label='Third Best path, lenght = '+ str(lenght_third ))   
plt.xlabel('x')
plt.ylabel('y')
plt.title("Disposizione Circolare ")
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True) 
plt.legend()

plt.show()
```
