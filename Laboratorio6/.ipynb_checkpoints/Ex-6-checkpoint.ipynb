
# <span style="color:blue">  Laboratorio 6 - Modello di Ising</span>

Il modello di Ising  è un *sistema di spin* {$s_i$} posizionati ai Nodi di un reticolo N-dimensionale, i quali possono assumere i valori discreti $± 1$. 

Gli **spin** interagiscono tra primi vicini, con costante di accoppiamento $J$; interagiscono inoltre con il campo magnetico esterno $h$. L'**Hamiltoniana** che descrive l' interazione è data da ( assumiamo che $\mu_B=1$ and $k_B=1$) :
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove si assume che valgano le *condizioni periodiche al contorno* $s_{N+1} = s_1$.

In meccanica statistica è quindi possibile definire un a **funzione di partizione** $Z$ (ci troviamo in un ensamble *Canonico*) associata il modello, attraverso la quale, esplorando lo spazio delle fasi, è possibile ricavare funzioni termodinamiche necessari allo studio del sistema, quali : energia interna per spin $U/N$, calore specifico $C$,suscettività magnetica $\chi$ e magnetizzzione $M$.

Il codice presente in `Ex-6/Monte_Carlo_ISING_1D.cpp` permette di simulare l'evoluzione di un sistema di spin, definito come un modello di Ising 1D, a temperatura $T$ costante.

Sono stati sviluppati  $2$ possibili algoritmi per il campionameto del **peso di Boltzmann**  $p(\{ s_i \}) = \exp(-\beta H)/Z$ ( il quale determina come avviene ciascun step verso l'equilibrio; nel codice è definito dalla funzione `Move`):
1. **campionamento di Metropolis**
2. **campionamento di Gibs**

Ne illustro ora il funzionamento, cosi come presente nella funzione `Move` all'interno del main:

#### 1.  Campionamento di Metropolis
- selezione Random di uno *spin* $s_a$.
- *flip* dello spin $s_a$.
- Calcolo di $\Delta E = 2Js_a(s_{a-1} + s_{a+1})+2hs_a$ ; ricordando di applicare le *boundary conditions* al problema.
- Lo scambio è accettato con probabilià $A = min[1,e^{-\beta \Delta E}]$, ovvero è accettato al $100$% nel caso in cui l' energia del sistema diminuisce, mentre una parte delle mosse che portano a configurazioni più improbabili sono rigettate.

#### 2.  Campionamento di Gibbs
- selezione Random di uno *spin i-esimo*.
- Assegno allo spin *i-esimo* un valore casuale tra {-1,1}, preso dalla distribuzione condizionata dei primi vicini.
- Calcolo la probabilità che lo spin i-esimo assuma uno dei due valori possibili:
$$p(s_{\pm 1}) = 1 / (1 + e^{-\beta \Delta E})$$
   con $\Delta E = 2Js_i(s_{i-1} + s_{i+1})+2hs_i$ ; ricordando di applicare le *boundary conditions* al problema.
- mediante la probabilità calcolata, impongo il valore dello spin calcolato nel *secondo punto* o impongo il *flip*
Il campionamento di Gibbs è più efficiente in quanto la mossa proposta viene sempre accettata.

###  Energia interna, Calore specifico, Suscettività
Di seguito sono mostrate le stime delle quantità termodinamiche, confrontate con la soluzione analitica.

Questa prima simulazione è stata svolta per `h`$=0$.

**Osservazione**:prima di passare da un sampling all' altro, è necessario lanciare i seguenti comandi da terminale:

2. make clean_dat

L' esecuzione del programma è poi lanciata mediante `./temp.sh`

Le simulazioni sono state fatte mediante con `nstep` $= 100000$ e `nblk` $= 20$, dopo aver eseguito una prima fase di *Equilibrazione*.

I file in cui sono presenti i risultati delle simulazioni sono in`Ex-6/output/<sampling type>`.

```python
plt.figure(figsize=(25,19))
plt.suptitle("Modello di Ising 1D", fontsize = 30)
points=100
Te = np.linspace(0.2,3.0,num=points)

#---------Energia Interna--------
plt.subplot(3,1,1)
    #Curva Esatta

beta = 1/Te
J = 1.0
Ns = 50
th = np.tanh(J/Te)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(Te, e,label = "Espressione analitica")

    #Metropolis
T, ene,err_ene = np.loadtxt("Ex-6/output/Metropolis/ene.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,ene ,yerr=err_ene,marker = 'o',markersize = 15,color = 'r',label = "Metropolis",linestyle = "none")

    #Gibbs
T, ene_g,err_ene_g = np.loadtxt("Ex-6/output/Gibbs/ene.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,ene_g ,yerr=err_ene_g,marker = 'o',markersize = 8,color = 'b',label = "Gibbs",linestyle = "none")

plt.title("Energia Interna, Modello di Ising 1D ", fontsize = 20)
plt.xlabel('T')
plt.ylabel('U/N')
plt.grid(True)
plt.legend()
             
#---------Calore Specifico--------
plt.subplot(3,1,2)
    #Curva Esatta
heat_a=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(Te, heat_a,label = "Espressione analitica")             

    #Metropolis
T, heat,err_heat = np.loadtxt("Ex-6/output/Metropolis/heat.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,heat ,yerr=err_heat,marker = 'o',markersize = 15,color = 'r',label = "Metropolis",linestyle = "none")

    #Gibbs
T, heat_g,err_heat_g = np.loadtxt("Ex-6/output/Gibbs/heat.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,heat_g ,yerr=err_heat_g,marker = 'o',markersize = 8,color = 'b',label = "Gibbs",linestyle = "none")

plt.title('Calore Specifico, Modello di Ising 1D ', fontsize = 20)
plt.xlabel('T')
plt.ylabel('C')
plt.grid(True)
plt.legend()

#---------Suscettività--------
plt.subplot(3,1,3)
    #Curva Esatta
    
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(Te, X,label = "Espressione analitica")       

    #Metropolis
T, chi,err_chi = np.loadtxt("Ex-6/output/Metropolis/chi.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,chi ,yerr=err_chi,marker = 'o',markersize = 15,color = 'r',label = "Metropolis",linestyle = "none")

    #Gibbs
T, chi_g,err_chi_g = np.loadtxt("Ex-6/output/Gibbs/chi.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,chi_g ,yerr=err_chi_g,marker = 'o',markersize = 8,color = 'b',label = "Gibbs",linestyle = "none")

plt.title('Suscettività, Modello di Ising 1D ', fontsize = 20)
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.grid(True)
plt.legend()
             
plt.show()
 
```

La compatibilità tra le misure e le previsioni teoriche è buona. Inoltre non noto una differenza significativa tra la bontà dei due Sampling.



### Magnetizzazione

La simulazione contiene la magnetizzazione, sia per `h` $= 0$( dove il valore medio dovrebbe essere nullo) , che per `h` $= 2$, per entrambi i campionamenti, svolta come sopra descritta

```python
plt.figure(figsize=(25,19))
plt.suptitle("Magnetizzazione, Modello di Ising 1D", fontsize = 30)
points=100
Te = np.linspace(0.2,3.0,num=points)

#------------------------------h = 0-------------------------
plt.subplot(1,2,1)
h=0.00 #external field
b = 1/Te
#Espressione analitica
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(Te, M,label = "Espressione analitica")

#Metropolis
T, m_0,err_m_0 = np.loadtxt("Ex-6/output/Metropolis/mag_0.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,m_0 ,yerr=err_m_0,marker = 'o',color = 'r',label = "Metropolis")

#Metropolis
T, m_0,err_m_0 = np.loadtxt("Ex-6/output/Gibbs/mag_0.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,m_0 ,yerr=err_m_0,marker = '.',color = 'b',label = "Gibbs")

plt.title('Ising 1D, magnetization M with h = 0.00')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()
plt.grid()

#------------------------------h = 0.02-------------------------
plt.subplot(1,2,2)
h=0.02 #external field
b = 1/Te
#Espressione analitica
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(Te, M,label = "Espressione analitica")

#Metropolis
T, m_2,err_m_2 = np.loadtxt("Ex-6/output/Metropolis/mag_2.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,m_2 ,yerr=err_m_2,marker = 'o',markersize = 15,color = 'r',label = "Metropolis",linestyle = "none")

#Gibbs
T, m_2,err_m_2 = np.loadtxt("Ex-6/output/Gibbs/mag_2.dat",usecols = (0,1,2), delimiter = None, unpack = 'true')
plt.errorbar(T,m_2 ,yerr=err_m_2,marker = 'o',markersize = 8,color = 'b',label = "Gibbs",linestyle = "none")

plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()
plt.grid()


plt.show()

```

Le oscillazioni della suscettivita $M_{h = 0.02}$ attorno a $T = 0.5$ sono ampie, ma anche gli errori sono maggiori, e dunque i risultati sono comunque compatibili con le previsioni teoriche.

**Osservazione generale** :A temperature basse, per tutti i grafici plottati, è infatti possibile notare maggiori *discostamenti*; questo potrebbe essere riconducibile al fatto che per $T\rightarrow 0$ il sistema perde di ergodicità, quindi la regione dello spazio delle fasi esplorabile dal Metropolis si riduce e dunque l' algoritmo diminuisce in efficienza.
