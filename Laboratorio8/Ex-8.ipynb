Il Variational Monte Carlo è una tecnica che permette di ottenere un'approssimazione dell'energia e della funzione d'onda del ground state di un sistema many-body ottimizzando una famiglia di funzioni d'onda {$\Psi_T(x)$} , rispetto a una serie di parametri che definiscono la famiglia, per ottenere l'energia minima. Sfrutta il principio Variazionale della MQ secondo il quale: 
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

dove $E_0$ è l'energia del ground state $\Psi_0$.

Il modello è stato applicato, all'interno del notebook, a una **particella** confinata in uno spazio 1D dal seguente potenziale:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

Per ottenere una approssimazione della funzione d'onda del ground state, il *VMC method* usa una funzione d'onda di prova, parametrizzata da un set di **parametri variazionali**. Considerando la simmetria del problema e potenziale, è stata scelta come funzione d'onda la seguente: 
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

la quale ha $2$ parametri variazionali : $\mu$ e $\sigma$. 

La ricerca del minimo di $ \langle{\hat H}\rangle_T$ avverà nello spazio dei $2$  parametri variazionali.

Di seguito sono illustrate due metodi differenti per trovare il minimo di $ \langle{\hat H}\rangle_T$.

## Exercise 08.1 
### Campionamento dell'integrale mediante Metropolis 
All' interno dell' esercizio è stato utilizzato l'**algoritmo di Metropolis** per campionare la densità di probabilità della funzione d'onda di prova  $|\Psi_T^{\sigma,\mu}(x)|^2$, usando una transizione di probabilità *uniforme* $T(x_{new}|x_{old})$.

L'obiettivo è quello di calcolare, per un dato $\mu_{test}$ e $\sigma_{test}$ fissato, il valore di 
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

usando il *Data-Blocking* ($\hbar = 1$ and $m=1$), per poi arrivare ai valori $\mu_{best}$ e $\sigma_{best}$ che minimizzano l'energia.

Per minimizzare il valore di $\langle {\hat H} \rangle_T$, è necessario fissare dei valori iniziali di $\mu_{in}$ e $\sigma_{in}$ dal quale la simulazione parte. I valori sono fissati all' interno del file  `input.dat `. Mediante l'utilizzo di **Geogebra** ho studiato quale valori meglio approssimano la funzione analitica fornita, almeno in termini di forma. 

![](Grafico_in.jpg)

Ho scelto quindi di utilizzare come parametri iniziali: $\mu_{in} = 0.90$ e $\sigma_{in} = 0.65$.

### Funzionamento del Codice

All' file `input.dat ` deve essere selezionato `SA`$=0$, sono inseriti i valori di `mu` e `sigma ` e la larghezza degli step con cui questi vengono modificati, oltre alle variabili calssiche per il Data-Blocking.

Analizzo ora il funzionamento delle **funzioni** nel main
- `Input`: carica i dati presenti nel file `input.dat `e inizializza il problema.

- `Averages `: calcola il valore di $\langle {\hat H} \rangle_T$ all'interno di ogni blocco, dove vale : 

$$
\langle {\hat H} \rangle_T = \frac{{-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}} \Psi_T^{\sigma,\mu}(x)}{\Psi_T^{\sigma,\mu}(x)}
$$ Ogni step di MonteCarlo è fatto mediante Metropolis mediante la funzione `Move_MC_metro`. In ogni  step, a partire dalla posizione $x_{old}$, la funzione  valuta la relazione tra la distribuzione di probabilita valutata in $x_{old}$ e $x_{new}$ e statisticamente decide se accettare la nuova posizione.

Mediante degli **istogrammi delle posizioni campionate**, è possibile valutare la coerenza tra la simulazione, la funzione d'onda di prova e la soluzione analitica. Gli istogrammi mostrati sono relativi esclusivamente alle casistiche in cui `nblk `$=50$.

```python
#===== funzione d'onda di prova======
histofill_blk = 1000                  # numero di posizioni salvate per blocco (cfr input.dat)
x = np.linspace(-2,2,100)      # uso l'ordinamento da 1 a N

mu = 0.9
sgm = 0.65
def f(x):
   return np.e**((-(x-mu)**2)/(2*sgm**2)) + np.e**((-(x+mu)**2)/(2*sgm**2))

#===== Soluzione analitica ========
def Vpot(x):
    return (x**2 - 2.5)*x**2
hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
# Step sizes
x_a = np.linspace(-a/2, a/2, N)
dx = x_a[1] - x_a[0] # the step size
V = Vpot(x_a)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)
# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

#======= plotting istogramma========
plt.figure(figsize = (12,8))
plt.suptitle("Campionamento delle posizioni")

plt.subplot(1,2,1)
pos_2 = np.loadtxt("Ex-8/8.1/pos_2.out", usecols = (0), delimiter = ' ', unpack = 'true')
n_bins = 200
x = np.linspace(-3,3,100)
plt.hist(pos_2, n_bins,density = 'true',label= '$\mu=1.0$ e $\sigma=0.5$')
plt.xlabel('Posizione')
plt.ylabel('Frequenza')
plt.grid(True)
mu = 1
sgm = 0.5
plt.plot(x,f(x)**2/2,label = 'Funzione prova')
plt.plot(x_a,(psi[0])**2,label = 'Funzione analitica',color = 'r')
plt.grid(True)
plt.xlim((-3,3))
plt.legend()

plt.subplot(1,2,2)
pos_1 = np.loadtxt("Ex-8/8.1/pos_1.out", usecols = (0), delimiter = ' ', unpack = 'true')
n_bins = 200
x = np.linspace(-3,3,100)
plt.hist(pos_1, n_bins,density = 'true',label= '$\mu=0.90$ e $\sigma=0.65$')
plt.xlabel('Posizione')
plt.ylabel('Frequenza')
plt.grid(True)
mu = 0.90
sgm = 0.65
plt.plot(x,(f(x)**2)/2,label = 'Funzione prova')
plt.plot(x_a,(psi[0])**2,label = 'Funzione analitica',color ='r')
plt.grid(True)
plt.xlim((-3,3))
plt.legend()


plt.show()
```

Coerentemente a quanto mi sarei aspettato, usando valori di $\sigma$ e $\mu$ più vicini a quelli che minimizzano l' energia, ottengo un grafico dell' energia pressochè costante , il cui valore si avvicina a quello del *Ground State* calcolato analiticamente. Inoltre anche l'istogramma è coerente con la distribuzione teorica.

## Exercise 08.2
### Simulated Annealing

Il Simulated Annealing consiste nell'associare a qualche parametro scelto una mobilità, che tipicamente si quantifica come una temperatura. Il sistema studiato viene quindi fatto evolvere facendo calare la "temperatura" affinche il parametro possa visitare le configurazioni accessibili e fermarsi in un minimo. Nel caso in studio, i parametri saranno $\sigma$ e $\mu$.

### Funzionamento del codice
All' file `input.dat ` deve essere selezionato `SA`$=1$, sono inseriti i valori di `mu` e `sigma ` e la larghezza degli step con cui questi vengono modificati, oltre alle variabili calssiche per il Data-Blocking.

Il processo di SA è stato implementato all' interno del `main`: esso consiste di `nstepSA` $=50$ step, ad ogni step si verifica un abbassamento della *temperatura*.

**Osservazione**:Sarebbe stato anche  possibile far si che il ciclo di abbassamento si fermi una volta che il valore di $\mu$ e $\sigma$ è pressoche costante. Ho deciso di non adottare questo metodo ma di calibrare il numero di volte in cui viene abbassata la temperatura, studiando l'andamento dei parametri.

La **temperatura** iniziale del simulated annealing è stata fissata a $T = 1$ e viene aggiornata secondo la seguente:
$$\beta = \beta_{old} + d\beta$$

dove $\beta = 1/T$ e $d\beta = 1$ ( come impostato nel file `input.dat`).

**Osservazione**: per ciascuna temperatura analizzata, è stato svolto un sampling mediante Metropolis, in maniera analoga all' esercizio *08.1*.

In `Ex-8.cpp` è stato implementato del codice in modo tale che vengano salvati su file i valori di *Energia* e *Posizioni* rigurdanti solamente l'ultimo step Metropolis, per ciascuna *temperatura* del simulated Annealing.

Per ogni *temperatura*, è stata svolta una fase di **Equilibrazione** per avvicinarmi al punto di Minimo.
