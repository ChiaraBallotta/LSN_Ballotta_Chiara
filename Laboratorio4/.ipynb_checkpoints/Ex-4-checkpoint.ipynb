```python
import numpy as np
import random
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
```

Per l'esercitazione, è stato fornito del codice che, mediante tecniche di **Dinamica Molecolare** *(MD)* è in  grado di calcolare l'evoluzione dinamica di un sistema di $N$ particelle e di misurarne:
- Energia interna totale per particella $E/N$
- Energia potenziale per particella $U/N$
- Energia cinetica per particella $K/N$
- Temperatura $T$

L'evoluzione dinamica è stata fatta nell' ipotesi di trattare con un **Ensamble Microcanonico**. Dal punto di vista *macroscopico* questo ensemble rappresenta i possibili stati di un sistema la cui energia totale $E$ è fissata; dal punto di vista *microscopico* l'ensemble microcanonico è invece definito dal postulato di equiprobabilità, che permette di assegnare ad ogni microstato con energia prossima ad $E$ la medesima probabilità di essere visitato.

La Dinamica Molecolare è una  tecnica di simulazione introdotta per risolvere uno specifico problema di meccanica statistica: ovver studiare l' evoluzioni di variabili macroscopiche di un sistema.Noto il potenziale di interazione, è possibile simulare analiticamente la soluzione dell' Hamiltoniana che descrive il sistema, con opportune condizioni al contrno; da esse è quindi possibile ricavare informazioni sulle grandezze mascroscopiche, mediante l'uso delle medie di Ensamble. 

La risoluzione dell' Hamiltoniana, e quindi il calcolo delle equazioni differenziali annesse è svolto mediante l'**Algoritmo di Verlet**.

Il codice fornito permette di studiare la dinamia molecolare di *Solidi*, *Gas*,*Liquidi*; in base alla fase del materiale, è necessario utilizzare un file di Input specifico, che definisce i parametri della simulazione. I file di input sono presenti in `Ex-4\input`. All'interno del **file di input** è possibile definire:
- `temp` : temperatura. avrà un valore differente a seconda che ci si trova nella fase di *Equilibrazione* o di *Evoluzione Dinamica*

- `npart` : numero di particelle

- `rho`

- `rcut` : raggio di *Cut-off*, ovvero la distanza oltre la quale approssimiamo nulla l'interazione tra coppie di particelle

- `delta ` : passo dell' algortimo di **Verlet**, usato per integrare le equazioni del moto di Newton.

- `Restart` il cui valore può essere 0 o 1. Il suo valore è collegato al processo di *Equilibrizzazione*.

- `nstep` : numero di step con cui viene fatto evolvere il sistema

## Equilibrizzazione

Le prime posizioni date all' algoritmo di Verlet, estratte dal file `Ex-4/Input/Config.in` non sono adatte a descrivere il sistema alla `temp` prescelta: descrivono un reticolo cristallino fcc alla temeratura $T=0$; la configurazione iniziale necessita del "tempo" affinchè possa "sciogliersi" e assumere il comportamento del fluido in studio.

Per questo , prima di iniziare lo studio dell' **Evoluzione Dinamica** del sistema, è necessario equilibrare il sistema facendolo evolvere verso una condizione di equilibrio.

Lo studio della dinamica molecolare si divide quindi in due parti:
- `restart = 0` : fase di *equilirizzazione*. La configurazione iniziale è estratta dal file `Config.in`.La temperatura, dopo una serie di step, converge alla temperatura $T_{eq}$, attorno al quale si stabilizza. La temperatura di input $temp$ viene  quindi modificata in modo tale che la temperatura a cui converge il sistema in equilibrio sia $T^\star$. L'ultima configurazione ( il sistema è in equilibrio) è salvata su `Config.out`


- `restart = 1` : fase di *dinamica molecolare*, in cui la temperatura rimane stabile attorno a $T^\star$.La configurazione iniziale è estratta dal file `Config.out`

Durante questa fase è stata studiata la *Temperatura*, ma qualsiasi altra grandezza macroscopica sarebbe stata equivalente.

In seguito sono illustrati i grafici utilizzati per capire se il `nblk` utilizzati sono sufficienti affinchè la temperatura si stabilizzi attorno al valore prefissato $T^\star$.  Il numero degli step per ciascun blocco è stato fissato a 1 : `nstep` = 1.

Le temperature Target $T^\star$ per le varie fasi  sono le seguenti : 
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)
 
Per ciascuna fase ho fissato `temp` = $T^\star$ e ho verificato che per gli `nblk` selezionati, le $3$ fasi raggiungono una condizione stazionaria,oscillando attorno alla temperatura $T_{eq}$.

Ho quindi modificato `temp` in modo tale che $T_{eq}$ = $T^\star$. Per ottenere il risultato desiderato sono andata per tentativi, incrementando e aumentando di poco la  `temp`.

Questo passaggio è fondamentale  in quanto sono interessata a studiare la *Dinamica Molecolare* alla temperatura scelta $T^\star$ e non a una casuale.

I risultati ottenuti durante la fase di Equilibrazione sono i seguenti ( ho indicato anche i `nblk` che ho reputato necessari fare per arrivre a una condizione di stabilità):
- **Fase Solida** : `temp` $=  1.54$ ; `nblk` $= 10000$
- **Fase Liquida** : `temp` $=  1.96$ ; `nblk` $= 10000$
- **Fase Gassosa** : `temp` $=  1.01$ ; `nblk` $= 14000$

I dati ottenuti durante la fase di *Equilibrazione* sono salvati nei file  `Dati_sim/<Fase>/simulaz_eq.dat`

```python
temp_target = [0.8 , 1.1 , 1.2]
Fase = ["Solid", "Liquid", "Gas"]
ave = np.array([0.0,0.0,0.0])

plt.figure(figsize=(21,15))
plt.suptitle("Equilibrazione Temperatura", fontsize = 25)

#plotting delle simulazioni, relative alle tre fasi

for i in range (3) :
    plt.subplot(3,1,i+1)
    N, temp = np.loadtxt("Ex-4/Dati_sim/"+ Fase[i] + "/simulaz_eq.dat",usecols = (0,1), delimiter = None, unpack = 'true')
    ave[i] = np.mean(temp)
    plt.plot(N,temp ,label=Fase[i] + " Phase")
    plt.title(Fase[i] + " Phase", fontsize = 20)
    plt.xlabel('Step number')
    plt.ylabel('Temprature')
    plt.axhline(y = temp_target[i], color ="r", label = "Target temperature: "+str(temp_target[i]))
    plt.axhline(y = ave[i], color ="darkorange", label = "Equilibration Temperature: "+str(ave[i]), linestyle = ':')
    plt.grid(True)
    plt.legend()
    
    plt.subplots_adjust(hspace=0.5) #separa la griglia di grafici
plt.show()      
```

L' equilibrazione risulta efficace per tutte e tre le fasi, sebbene la fase gassosa richieda più step per raggiungere la stabilità( coerente con il fatto che la fase gassosa ha una densità minore e dunque la redistribuzione dell'energia cinetica, che avviene mediante gli urti, sarà più lenta)

Durante la fase *iquida* e *solida* la temperatura diminuisce drasticamente, questo deriva dal fatto che il sistema parte da un sistema cristallino estremamente ordinato ma man mano che il sistema si disordina, l' energia potenziale aumenta( non sono più nel minimo) e per conservaziìione dell' energia, l' energia cinetica e quindi la temperatura devono diminuire.

Al contrario,nella *fase gassosa* il comportamento è opposto( l'energia potenziale diminuisce).Questo fenomeno potrebbe essere causato dal fatto che il cristallo fcc è molto lontano dalla configurazione di un gas ideale, dove domina l' energia cinetica.

Per ogni fase è stata calcolata la temperatura media ottenuta durante la fase di equilibrizzazione, per assicurarsi che $T_{eq} = T^\star$

```python
print("Temperatura media ottenuta mediante l'equilibrazione, per le varie fasi")
for i in range (0,3):
    print("     -" + Fase[i]+" phase : " + str(ave[i]))
```

Le temperature ottenute sono quindi state inserite nel file di Input, per svolgere le simulazioni di *Dinamica Molecolare* : `temp` = $T_{eq}$

## Exercise 04.1
 Il codice è stata aggiunta la logica per poter calcolare la pressione $P$, come medie di blocco, medie progressive e incertezze progressive.

La pressione è stata calcolata ricordando che, per il **Teorema del Viriale**,  la pressione in un sistema con potenziale di Lennard-Jones vale: 

$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

All' interno del *Notebook* sono state utilizzate le unità di misura ridotte.
